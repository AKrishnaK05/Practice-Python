```python
import matplotlib.pyplot as plt
```

```python
# PIE CHART

dept = ["HR","Sales","Technical","Service"]
cnt = [23,17,34,21]

plt.pie(cnt,labels=dept,autopct="%1.2f%%")
plt.show()
```

```python
# Using additional features
labels = ["ML","DS","AI","POWERBI"]
sizes = [40,50,80,30]
colours = ["lightblue","green","yellow","red"]
explode = [0,0,0.2,0]
plt.pie(sizes,labels=labels,colors=colours,explode=explode,autopct="%1.2f%%",shadow=True,startangle=90)
plt.show()
```

```python
# Adding styling

plt.title("Pie Chart")
slices = [32324,45346,34534,45346,34535]
labels = ["Java","Python","C","C++","SQL"]
explode = [0,0.2,0,0,0]
plt.pie(slices,labels=labels,shadow=True,startangle=90,explode=explode,autopct="%1.1f%%",wedgeprops={'edgecolor':'black','linewidth':2},colors=colours)
plt.tight_layout
plt.show()
```

```python
# LINE CHART
x = [1,2,3,4,5,6]
y = [4,5,6,3,2,3]

plt.plot(x,y,color="green")
plt.title("x vs y")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
