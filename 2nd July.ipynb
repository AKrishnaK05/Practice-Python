```python
# Lineplot

data = {'Time': [1, 2, 3, 4, 5], 'Value': [3, 5, 2, 4, 6]}
df = pd.DataFrame(data)

sns.lineplot(x='Time', y='Value', data=df)
plt.show() 
```
